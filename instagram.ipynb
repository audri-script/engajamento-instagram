```python
# tratando a base de tags, separando elas:
base.Tags = base.Tags.str.split('/')
base = base.explode('Tags')
base.head()
```

```python
# analisando as tags:
base.groupby('Pessoas')['Curtidas'].mean()
# como adicionamos novas colunas, A MEDIA DE QUALQUER OUTRA COLUNAS ALEM DA TAGS MUDA
# esse valores gerados estao errados
base.head()
# APOS SPLIT E EXPLODE, SO PODEMOS ANALISAR AS COLUNAS QUE GERAMOS (NESSE CASO, TAGS)
```

```python
base.groupby('Tags')[['Curtidas','Comentários']].mean().sort_values('Curtidas',ascending=False)
```

```python
# tratando os valores sem tag NaN:
base[base.Tags.isnull()]
```

```python
# transformando o NaN em N:
base.loc[base.Tags.isnull(),'Tags'] = 'Sem tag'
```

```python
# visualizando a media das sem tags:
base.groupby('Tags')[['Curtidas','Comentários']].mean().sort_values('Curtidas',ascending=False)
# podemos ver que mesmo sem tag, o engajamento nao eh ruim
```

```python
# se quisermos ignorar as sem tags, podemos voltar ao NaN:
base.loc[base.Tags == 'Sem tag','Tags'] = np.nan
base[base.Tags.isnull()]
```

```python
base.groupby(['Pessoas','Tags'])[['Curtidas','Comentários']].mean().sort_values('Curtidas',ascending=False)
# vemos que Pessoas eh o principal fator para ter um bom engajamento, principalmente em Produtos e Novos Produtos
```

```python
base.groupby(['Campanhas','Tags'])[['Curtidas','Comentários']].mean().sort_values('Curtidas',ascending=False)
# Influenciadores aparentam ser efetivos em Campanhas para o engajamento
# promocoes e trends mostram alto engajamento
# Pessoas sao essenciais para o engajamento, o rosto faz diferenca
```
